### Question 6 - Bonus: Put all the previously mentioned data transformations into a function

```python
def transformed_data(df):
    df.columns = map(lambda x: x.lower().replace("-", "_").replace(" ", "_"), df.columns) # lowering headers
    df['state'].fillna("Unknown", inplace=True)
    df['response'].fillna("N/A", inplace=True)
    df['months_since_last_claim'].fillna(1, inplace=True)
    df['number_of_open_complaints'].fillna(1, inplace=True)
    df['vehicle_class'].fillna('Unknown', inplace=True)
    df['vehicle_size'].fillna('Unknown', inplace=True)
    df['vehicle_type'].fillna('Unknown', inplace=True)
    df['effective_to_date'] = pd.to_datetime(df['effective_to_date'], errors='coerce')
    df['month'] = df['effective_to_date'].dt.month
    df[df['month']<4]
    return df

transformed_data(df)
```
